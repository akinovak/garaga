```python
def is_pth_residue(x): 
    return x^(h3 * 27) == ONE

def get_pth_root_inverse(x): 
    if is_pth_residue(x): 
        return ONE 
    
    v = 27 * h3 
    wj = x^v 
    
    v_inv = inverse_mod(v, p)
    s = (-1 * v_inv) % p
    
    return wj^s
```

```python
def get_order_of_3rd_primitive_root(x): 
    # correct way is do do r * p * h3 but outputs of equal Tate pairings are always of the form c^r thus there is no rth root contribution
    y = x^(p * h3) 
    
    if y == ONE: 
        return 0 
    
    if y^3 == ONE: 
        return 1
    
    if y^9 == ONE: 
        return 2 
    
    if y^27 == ONE: 
        return 3
    

def get_any_27th_root_inverse(x): 
    pw = get_order_of_3rd_primitive_root(x)
    
    if pw == 0: 
        return ONE
    
    _ord = 3^pw
    
    v = p * h3 
    wj = x^v 
    
    v_inv = inverse_mod(v, _ord)
    s = (-1 * v_inv) % _ord
    
    return wj^s
```

```python
def h3_ord_element_lambda_root(x): 
    # after appying shifts we know that element is order just h3 
    
    e = inverse_mod(lam, h3)
    return x^e
```

```python
for i in range(100): 
    f = Fq12.random_element()
    f = f^r # it will always come as r-th residue in the honest execution

    wp_shift = get_pth_root_inverse(f)
    w27_shift = get_any_27th_root_inverse(f)
    
    f_shifted = f * wp_shift * w27_shift

    root = h3_ord_element_lambda_root(f_shifted)
    assert(f_shifted, root^lam)
    print('step: ', i)
```
